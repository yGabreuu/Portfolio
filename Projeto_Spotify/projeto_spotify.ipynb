ID : identificador exclusivo para cada faixa no Spotify.

Name : Título da faixa.

Popularity : Índice de popularidade do Spotify para a faixa, indicativo de sua frequência de audição.

5. Visualização de dados

```python
principais_faixa = spotify_df.sort_values('Popularidade',ascending=False)
melhores_musicas = principais_faixa.head(10)
melhores_musicas
```

5.1-Gráfico feito para saber quais são as 10 melhores músicas com mais popularidade

```python
fig, ax = plt.subplots(figsize=(10,5))
ax.bar(melhores_musicas['Músicas'], melhores_musicas['Popularidade'], label='red', color='r')
labels = ax.get_xticklabels()
plt.setp(labels,rotation=90)
ax.set_xlabel(xlabel='TOP 10 MÚSICAS')
ax.set_ylabel(ylabel='Popularidade')
```

```python
principais_faixa = spotify_df.groupby('Músicas')[['Popularidade']].sum().reset_index()
principais_faixa = principais_faixa.sort_values('Popularidade', ascending=False)
dez_melhores_faixas = principais_faixa.head(10)
dez_melhores_faixas
```

```python
grupo_nome = principais_faixa['Músicas'].head(10)
grupo_data = principais_faixa['Popularidade'].head(10)
grupo_data.mean()
```

5.2-Gráfico para saber quais são as 10 músicas mais procurada.

```python
fig , ax = plt.subplots(figsize=(14,5))
cores = ['r' if num >= 200
         else 'b' for num in grupo_data.sort_values(ascending=False)]
ax.bar(grupo_nome,grupo_data, label='blue', color=cores)
labels = ax.get_xticklabels()
plt.setp(labels,rotation=45)
ax.set(xlabel='10 melhores músicas mais tocada', ylabel='Contagem')
```

5.3-Gráfico usado para saber quais sao os 5 melhores artista do gênero HINDI

```python
hind = spotify_df.loc[spotify_df['Músicas']=='Hindi']
hind_artista = hind.groupby('Artistas')[['Popularidade']].sum()
melhores_artistas_hindi = hind_artista.sort_values(by='Popularidade', ascending=False).reset_index()
melhores_artistas_hindi = melhores_artistas_hindi.head(5)
melhores_artistas_hindi
```

```python
yr = sns.catplot(data=melhores_artistas_hindi, x=melhores_artistas_hindi['Artistas'], 
            y=melhores_artistas_hindi['Popularidade'],
            errorbar=('pi',95),
            kind='bar',
            palette='Reds_r')
yr.set_xticklabels(rotation=45)
yr.set(xlabel='Melhores artistas HINDI',ylabel='Popularidade')
```

5.4- Gráfico usado para saber quais as músicas que contém inscrição remix, True="tem inscrição remix",False="não tem inscrição remix".

```python
# Count the occurrences of "Remix" in the Name column
spotify_df['Remix'] = spotify_df['Músicas'].apply(lambda x: 'Remix' in x)

# Count the number of entries with and without "Remix"
remix_counts = spotify_df['Remix'].value_counts()
remix_counts
```

```python
cores_1 = ['b' if cor >= 871 else 'y' for cor in remix_counts.sort_values(ascending=False)]
plt.figure(figsize=(8,5))
plt.pie(remix_counts, labels=remix_counts.index, 
        autopct="%1.1f%%"
        ,startangle=100
        ,colors=cores_1)
plt.title('Distribuição de inscrições com e sem "Remix"')
plt.axis('equal')
plt.show()
```
